# Clustering of Student Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv('turkish.csv')
#Lets cluster them on the basis of questions
data_questions=data.iloc[:,5:33]

# Dimensionality Reduction
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
data_questions_pca=pca.fit_transform(data_questions)

from sklearn.cluster import KMeans
wcss=[]
for i in range(1,7):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(data_questions_pca)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,7),wcss)
plt.title('The Elbow Method')
plt.xlabel('The number of clusters')
plt.ylabel('WCSS')
plt.show()

kmeans = KMeans(n_clusters = 3)
y_kmeans = kmeans.fit_predict(data_questions_pca)

kmeans = KMeans(n_clusters = 3)
y_kmeans = kmeans.fit_predict(data_questions_pca)


# Visualising the clusters
plt.scatter(data_questions_pca[y_kmeans == 0, 0], data_questions_pca[y_kmeans == 0, 1], s = 100, c = 'yellow', label = 'Cluster 1')
plt.scatter(data_questions_pca[y_kmeans == 1, 0], data_questions_pca[y_kmeans == 1, 1], s = 100, c = 'pink', label = 'Cluster 2')
plt.scatter(data_questions_pca[y_kmeans == 2, 0], data_questions_pca[y_kmeans == 2, 1], s = 100, c = 'red', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'blue', label = 'Centroids')
plt.title('Clusters of students')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```
